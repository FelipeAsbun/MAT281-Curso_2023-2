```python
import pandas as pd
```

```python
# load data
url='https://drive.google.com/file/d/1SW1_vWBLd50COj4FbQdZenqAwA0nLDqC/view?usp=drive_link'
url='https://drive.google.com/uc?id=' + url.split('/')[-2]

df = pd.read_csv(url, sep="|" )
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1.- ¿Cuál es el número de observaciones en el conjunto de datos?

```python
n_col = df.shape[1]
print(f"Número de columnas: {n_col}")
```

7.- Imprimir solo la columna de **occupation**.

```python
df["occupation"]
```

8.- ¿Cuántas ocupaciones diferentes hay en este conjunto de datos?

```python
n_occupation = df["occupation"].nunique()
print(f"Hay {n_occupation} ocupaciones distintas")
```

9.- ¿Cuál es la ocupación más frecuente?

```python
most_freq_occupation = df["occupation"].value_counts().idxmax()
print(f"La ocupación más frecuente es {most_freq_occupation}")
```

10.- ¿Cuál es la edad media de los usuarios?

```python
mean_age = round(df["age"].mean())
print(f"La edad media de los usuarios es de {mean_age} años")
```

11.- ¿Cuál es la edad con menos ocurrencia?

```python
least_freq_age = df["age"].value_counts().idxmin()
print(f"La edad de menor frecuencia es de {least_freq_age} años")
```
