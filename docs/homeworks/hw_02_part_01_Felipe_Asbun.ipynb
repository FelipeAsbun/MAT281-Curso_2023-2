Modificando prod:
- Sectors contiene datos con más de un tipo de sector. Codificaremos sus valores en 3 columnas nuevas, una por cada sector.
- 'primary_essential_function' contiene 2 niveles de información. También los separaremos en dos columnas.

```python
serie1 = prod['sectors']
serie2 = prod['primary_essential_function']

l_prek = []
l_high = []
l_corp = []
for i, s in enumerate(serie1):
    if not pd.isnull(s):
        if 'PreK-12' in s:
            l_prek.append(1)
        else:
            l_prek.append(0)
        if 'Higher Ed' in s:
            l_high.append(1)
        else:
            l_high.append(0)
        if 'Corporate' in s:
            l_corp.append(1)
        else:
            l_corp.append(0)
    else:
        l_prek.append(0)
        l_high.append(0)
        l_corp.append(0)

prod.drop(columns='sectors', inplace=True)
prod['PreK-12'] = l_prek
prod['Higher_Ed'] = l_high
prod['Corporate'] = l_corp


l_func = []
l_subfunc = []
for i, s in enumerate(serie2):
    if not pd.isnull(s):
        l_func.append(s.strip().split('-')[0].strip())
        l_subfunc.append(s.strip().split('-')[1].strip())
    else:
        l_func.append(np.nan)
        l_subfunc.append(np.nan)
prod.drop(columns='primary_essential_function', inplace=True)
prod['primary_essential_function'] = l_func
prod['primary_essential_subfunction'] = l_subfunc

```
