# Preparaci√≥n de los datos

Como 'Cabin' y 'Embarked' son datos categ√≥ricos, marcaremos los datos NaN como desconocidos, asign√°ndoles el valor 'Unknown'.

Para 'Age', reemplazaremos estos datos por la mediana. En 'Fare' reemplazaremos por la media.

````python
# Cantidad de datos nulos
display(X_train.isna().sum())
print('\n')
display(test.isna().sum())
````

Codificaci√≥n: Dado que es dif√≠cil sacar conclusiones directamente desde 'Name', 'Ticket', y 'Cabin', pues son muchos datos categ√≥ricos distintos, extraeremos patrones de estos. Adem√°s, del nombre podemos extrer otros elementos, como su largo o el n√∫mero de palabras en √©l.

Despu√©s, codificaremos los datos.

Para test haremos el mismo procedimiento, incluyendo lo anterior.

````python
# Preliminar
X_train['Cabin'] = X_train['Cabin'].fillna('Unknown')
X_train['Embarked'] = X_train['Embarked'].fillna('Unknown')
X_train['Age'] = X_train['Age'].fillna(X_train['Age'].median())
X_train['Fare'] = X_train['Fare'].fillna(X_train['Fare'].mean())

X_train['Class_ticket'] = X_train['Ticket'].str.extract('([a-zA-Z]+)', expand=False).fillna('Unknown')

X_train['Title'] = X_train['Name'].str.extract(' ([A-Za-z]+)\.', expand=False).fillna('Unknown')
X_train['Length_Name'] = [len(x.strip()) for x in X_train['Name']]
X_train['Name_words'] = [len(x.strip().split()) for x in X_train['Name']]

X_train['Floor'] = X_train['Cabin'].str.extract('([a-zA-Z]+)', expand=False).fillna('Unknown')

##

X_test['Cabin'] = X_test['Cabin'].fillna('Unknown')
X_test['Embarked'] = X_test['Embarked'].fillna('Unknown')
X_test['Age'] = X_test['Age'].fillna(X_train['Age'].median())
X_test['Fare'] = X_test['Fare'].fillna(X_train['Fare'].mean())

X_test['Class_ticket'] = X_test['Ticket'].str.extract('([a-zA-Z]+)', expand=False).fillna('Unknown')

X_test['Title'] = X_test['Name'].str.extract(' ([A-Za-z]+)\.', expand=False).fillna('Unknown')
X_test['Length_Name'] = [len(x.strip()) for x in X_test['Name']]
X_test['Name_words'] = [len(x.strip().split()) for x in X_test['Name']]

X_test['Floor'] = X_test['Cabin'].str.extract('([a-zA-Z]+)', expand=False).fillna('Unknown')
````

````python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

continuous_features = ['Age', 'Fare', 'Length_Name']
X_train_encoded = X_train[continuous_features]
X_test_encoded = X_test[continuous_features]

cont_scaler = StandardScaler()
cont_scaler.fit(X_train_encoded)

X_train_encoded = pd.DataFrame(cont_scaler.transform(X_train_encoded),columns=continuous_features)
X_test_encoded = pd.DataFrame(cont_scaler.transform(X_test_encoded),columns=continuous_features)

categorical_features=['Sex', 'SibSp', 'Parch', 'Embarked', 'Class_ticket', 'Title', 'Floor', 'Name_words']
cat_encoder = OneHotEncoder(sparse_output=False, handle_unknown='infrequent_if_exist')
cat_X_train = cat_encoder.fit_transform(X_train[categorical_features])
cat_X_train = pd.DataFrame(cat_X_train, columns=cat_encoder.get_feature_names_out(categorical_features))
cat_X_test = cat_encoder.transform(X_test[categorical_features])
cat_X_test = pd.DataFrame(cat_X_test, columns=cat_encoder.get_feature_names_out(categorical_features))

X_train_encoded = pd.concat([X_train_encoded, cat_X_train], axis=1)
X_test_encoded = pd.concat([X_test_encoded, cat_X_test], axis=1)
````

````python
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder

scores = []

for col in X_train_encoded.columns:
    score = mutual_info_classif(X_train_encoded[col].to_numpy().reshape(-1,1), y_train)[0]
    scores.append(score)

fig, axes = plt.subplots(figsize=(20,2), nrows=1, ncols=1)
plot_data = {'x': X_train_encoded.columns, 'y': scores}
sns.barplot(data=plot_data,x='x', y='y', hue = 'x', palette="Blues_d",ax=axes)
_=plt.xticks(rotation=90)
plt.show()
````

````python
from sklearn.model_selection import cross_val_score
from sklearn.svm import LinearSVC
from sklearn.svm import SVC

C_values = np.logspace(-4,2, num=7, endpoint=True, base=10.0)

test_accuracy = []
train_accuracy = []
n_supp_vect = []
cross_val_accuracy = []
cross_val_std = []

models = []

for C_val in C_values:
    svm = LinearSVC(C=C_val, max_iter=1000000)
    # svm = SVC(kernel='linear', C=C_val, max_iter=1000000)
    cv_scores = cross_val_score(svm, X_train_encoded, y_train, cv=10, scoring='accuracy')
    cross_val_accuracy.append(cv_scores.mean())
    cross_val_std.append(cv_scores.std())

    svm.fit(X_train_encoded, y_train)
    train_accuracy.append(svm.score(X_train_encoded, y_train))
    test_accuracy.append(svm.score(X_test_encoded, y_test))
    models.append(svm)
    n_supp_vect.append(len(np.where(np.abs(svm.decision_function(X_train_encoded))<=1)[0]))
    print('Done C:', C_val)

cross_val_accuracy = np.array(cross_val_accuracy)
cross_val_std = np.array(cross_val_std)

C_max = C_values[np.argmax(cross_val_accuracy)]

print('C_max:', C_max)
````

````python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(np.log10(C_values), train_accuracy, color='g', label='train_acc')
ax.plot(np.log10(C_values), test_accuracy, color='b', label='test_acc')
ax.plot(np.log10(C_values), cross_val_accuracy, color='r', label='cv_acc')
plt.fill_between(np.log10(C_values), cross_val_accuracy-cross_val_std, cross_val_accuracy+cross_val_std, alpha=0.2, color='gray')

fig.suptitle('Accuracy SVM Lineal vs C', size=10)
ax.legend(loc='center left', bbox_to_anchor=(1,0.9))
plt.show()
````

````python
n_supp_vect = np.array(n_supp_vect)

fs=10

fig, ax = plt.subplots(figsize=(8,5))
ax.plot(np.log10(C_values), n_supp_vect, color='b')
plt.title("Numero de support vectors versus valor de C",size=fs)
plt.ylabel('n_supp_vect')
plt.xlabel('C')
plt.show()
````

````python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

svm_linear = LinearSVC(C=C_max, loss='hinge', random_state=42, max_iter=1000000)
svm_linear.fit(X_train_encoded,y_train)

y_pred_train = svm_linear.predict(X_train_encoded)
y_pred_test = svm_linear.predict(X_test_encoded)
print('Train Set Report:')
print(classification_report(y_train, y_pred_train))
print('Test Set Report:')
print(classification_report(y_test, y_pred_test))

fig, ax = plt.subplots(figsize=(8,4), ncols=2, nrows=1)

cm_train = confusion_matrix(y_train, y_pred_train, normalize='true')
disp_train = ConfusionMatrixDisplay(confusion_matrix=cm_train, display_labels=[0,1])
disp_train = disp_train.plot(ax=ax[0],cmap=plt.cm.GnBu)
ax[0].set_title('Confusion Matrix Train Set', size=9)

cm_test = confusion_matrix(y_test, y_pred_test, normalize='true')
disp_test = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=[0,1])
disp_test = disp_test.plot(ax=ax[1],cmap=plt.cm.GnBu)
ax[1].set_title('Confusion Matrix Test Set', size=9)

fig.tight_layout()
plt.show()
````

````python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

C_values = np.logspace(-4,2, num=7, endpoint=True, base=10.0)
print(C_values)

test_accuracy_rbf = []
cross_val_accuracy_rbf = []
cross_val_std_rbf = []
train_accuracy_rbf = []
rbf_models = []

for C_val in C_values:
    svm = SVC(kernel='rbf', C=C_val, max_iter=10000)
    cv_output = cross_val_score(svm, X_train_encoded, y_train, cv=5)
    cross_val_accuracy_rbf.append(cv_output.mean())
    cross_val_std_rbf.append(cv_output.std())

    svm = SVC(kernel='rbf', C=C_val)
    svm.fit(X_train_encoded,y_train)
    train_accuracy_rbf.append(svm.score(X_train_encoded,y_train))
    test_accuracy_rbf.append(svm.score(X_test_encoded,y_test))
    rbf_models.append(svm)
    print("done C=", C_val)
````

````python
cross_val_accuracy_rbf = np.array(cross_val_accuracy_rbf)
cross_val_std_rbf = np.array(cross_val_std_rbf)
train_accuracy_rbf = np.array(train_accuracy_rbf)

C_rbf_max = C_values[np.argmax(cross_val_accuracy_rbf[(np.abs(train_accuracy_rbf - cross_val_accuracy_rbf) <= 2*cross_val_std_rbf).tolist()])]
print(C_rbf_max)

cross_val_accuracy_rbf = np.array(cross_val_accuracy_rbf)
cross_val_std_rbf = np.array(cross_val_std_rbf)

lw=1
fs=9

fig, ax = plt.subplots(figsize=(8,5))
ax.plot(np.log10(C_values),train_accuracy_rbf,color='g',lw=lw,label='train_acc')
ax.plot(np.log10(C_values),cross_val_accuracy_rbf,color='y',lw=lw,label='cv_acc')
plt.fill_between(np.log10(C_values), cross_val_accuracy_rbf-2*cross_val_std_rbf, cross_val_accuracy_rbf+2*cross_val_std_rbf,color='gray', alpha=0.2)
ax.plot(np.log10(C_values),test_accuracy_rbf,color='r',lw=lw,label='test_acc')
plt.title("Exactitud SVM RBF versus valor de C",size=fs)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.9),fontsize=fs)
plt.show()
````
